```python
# SVM Classification
import pandas as pd
import numpy as py

from sklearn import svm
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score

import matplotlib.pyplot as plt
```

```python
ff = pd.read_csv("forestfires.csv")
ff
```

```python
ff['size_category'].value_counts()
```

```python

# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
ff['size_category']= label_encoder.fit_transform(ff['size_category'])
ff['day']= label_encoder.fit_transform(ff['day']) 
ff['month']= label_encoder.fit_transform(ff['month'])
```

```python
ff
```

```python
x = ff.iloc[:,:-1]
x
y = ff.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=15)
```

# Grid Search CV

```python
clf = SVC()
param_grid = [{'kernel':["rbf"],'gamma':[50,5,10,0.5,1,0.001,0.0001,0.00001],'C':[1,15,14,13,12,11,10,0.1] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

# SVM Model

```python
#clf = SVC(c=15,kernel = "rbf", gamma=0.001)
clf = SVC(C= 14, gamma = 0.0001,kernel="rbf")
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred)*100
print("accuracy=",acc)
print(classification_report(y_test, y_pred))
```
