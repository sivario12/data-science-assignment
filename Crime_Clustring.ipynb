```python
#import hierarchical clustring libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

# Scaling

```python
# Normalization function
def norm_func(i):
    x = (i - i.min())/(i.max() - i.min())
    return x
```

```python
# Normalized data frame 
df_norm = norm_func(crime.iloc[:,1:])
```

```python
df_norm.iloc[:,:]
```

```python
#create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method = 'single'))
```

```python
??AgglomerativeClustring
```

```python
#create clusters
crimehc = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage = 'single')
```

```python
crime_hc=crimehc.fit_predict(df_norm)
```

```python
crime['h_clusterid'] = crime_hc
crime
```

```python
crime.groupby(["h_clusterid"])["Unnamed: 0"].apply(lambda x:tuple(x))
```

# Finding the optimal number of clusters

```python
data = pd.read_csv("crime_data.csv")
```

```python
nc = [1,2,3,4,5,6,7,8,9]
iner=[]
for i in nc:
    km=KMeans(i)
    km.fit(data.iloc[:,[1,2,3,4]])
    iner.append(km.inertia_)
```

```python
plt.plot(nc,iner,color="black")
plt.scatter(nc,iner,color="red")
plt.show()
```

# Silhouette_Score

```python
from sklearn.metrics import silhouette_score as sil
```

```python
nc=[2,3,4,5,6,7,8,9]
sc_score=[]    
for i in nc:
    hc = AgglomerativeClustering(n_clusters=i, affinity = 'euclidean', linkage = 'complete')

    y=hc.fit_predict(crime.iloc[:,[1,2,3]])
    sc_score.append(sil(crime.iloc[:,[1,2,3]],y))
```

```python
sc_score
```

```python
plt.plot(nc,sc_score,color="black")
plt.scatter(nc,sc_score,color="red")
plt.show()
```
