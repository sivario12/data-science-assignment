# Generl Recommendation

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings("ignore")
```

```python
pd.set_option("display.max_rows",500)
pd.set_option("display.max_columns",500)
```

```python
movies = pd.read_csv("my_movies.csv")
```

```python
movies.columns
```

```python
movies = movies[['SixthSense', 'Gladiator', 'LOTR1','HarryPotter1', 'Patriot', 'LOTR2', 'HarryPotter2', 'LOTR','Braveheart', 'GreenMile']]
```

```python
movies
```

```python
for i in movies.columns:
    print(i)
    print(movies[i].value_counts())
    print()
```

# Apriori Algorithm

```python
frequent_itemsets = apriori(movies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
movies[movies["Gladiator"]==1].shape[0]/movies.shape[0]
```

```python
rules = association_rules(frequent_itemsets,metric = "lift", min_threshold=1.3)
```

```python
rules
```

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold=1.1)
rules[rules["lift"]>1]
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
