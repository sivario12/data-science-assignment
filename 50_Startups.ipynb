# Preapring Stas Model for MLR

```python
import statsmodels.formula.api as smf
model = smf.ols("Profit~RandD_Spend+Administration+Marketing_Spend",data=startup).fit()
```

```python
model.summary()
```

```python
#claculating VIF(varience inflation factor) values 

rqs_rd = smf.ols("RandD_Spend~Administration+Marketing_Spend",data=startup).fit().rsquared
vif_rd = 1/(1-rqs_rd)
```

```python
rqs_ad = smf.ols("Administration~RandD_Spend+Marketing_Spend",data=startup).fit().rsquared
vif_ad = 1/(1-rqs_ad)
```

```python
rqs_ms = smf.ols("Marketing_Spend~Administration+RandD_Spend",data=startup).fit().rsquared
vif_ms = 1/(1-rqs_ms)
```

```python
d1 = {'variables':['RandD_SPend','Administration','Marketing Spend'],
     'VIf':[vif_rd,vif_ad,vif_ms]
     }
vif_frame=pd.DataFrame(d1)
vif_frame
```

# Test for Normality of Residuals

```python
plt.hist(model.resid,bins=20)
import statsmodels.api as sm
qqplot = sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

# Residual plot for homoscedasticity

```python
def get_standardized_values(vals):
    return(vals-vals.mean())/vals.std()

plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))
plt.title("Residual Plot")
plt.show()
```

# Residual VS Regressors

```python
fig = plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,"RandD_Spend",fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,"Administration",fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,"Marketing_Spend",fig=fig)
plt.show()
```

# Cook's Distance

```python
model_influence=model.get_influence()
(c,_)=model_influence.cooks_distance
k=3
n=50
3*(k+1)/n
```

```python
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(startup)),np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('cooks distance')
plt.show()
```
