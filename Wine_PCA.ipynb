# PCA(Principle Component Analysis)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
print(wine.shape)
print(x_pca.shape)
```

```python
x_pca
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=wine1['Type'],cmap='plasma')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
pca.components_
```

```python
finalDf = pd.concat([pd.DataFrame(x_pca[:,0:3],columns=['pc1','pc2','pc3']),wine1[['Type']]],axis=1)
finalDf
```

```python
df_comp = pd.DataFrame(pca.components_,columns=wine1.columns.values)
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```

```python
from sklearn.metrics import silhouette_score as sil
```

```python
nc=[2,3,4,5,6,7,8,9]
sc_score=[]
for i in nc:
    hc = AgglomerativeClustering(n_clusters=i,affinity = 'euclidean',linkage='complete')
    y = hc.fit_predict(finalDf.iloc[:,[1,2,3]])
    sc_score.append(sil(finalDf.iloc[:,[1,2,3]],y))
```

```python
sc_score
```

```python
plt.plot(nc,sc_score,color = "black")
plt.scatter(nc,sc_score,color="red")
plt.show()
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
TWSS = []
k = list(range(1, 6))

for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(finalDf)
    TWSS.append(kmeans.inertia_)
```

```python
TWSS
# Scree plot 
plt.plot(k, TWSS, 'ro-');plt.xlabel("No_of_Clusters");plt.ylabel("total_within_SS")
```

```python
# Selecting 3 clusters from the above scree plot which is the optimum number of clusters 
model = KMeans(n_clusters = 3) 
model.fit(finalDf)
```

```python
model.labels_ # getting the labels of clusters assigned to each row 
mb = pd.Series(model.labels_)  # converting numpy array into pandas series object 
finalDf['k-means'] = mb # creating a  new column and assigning it to new column
```

```python
finalDf
```

```python
kmeans = KMeans(n_clusters=3,random_state=0)
predict=kmeans.fit_predict(finalDf)
predict
```
